## Null Hypothesis (H₀)

The **null hypothesis (H₀)** is a default assumption that there is *no effect, no difference, or no relationship* between variables being studied. It represents the "status quo" — the position of skepticism that says "nothing interesting is happening here."

**Example:** If you're testing whether a new drug lowers blood pressure, H₀ would state: *"The drug has no effect on blood pressure."*

---

## Why It's Important

**1. It gives you something to test against**
You can't prove something *exists* directly — but you *can* try to disprove its absence. H₀ provides a concrete, falsifiable starting point. You gather evidence to see if reality is inconsistent with it.

**2. It controls false discoveries**
By requiring strong evidence to *reject* H₀ (typically p < 0.05), the framework guards against claiming effects that are just random noise.

**3. It anchors the p-value**
The p-value answers: *"If H₀ were true, how likely is it that we'd see data this extreme by chance?"* A very small p-value suggests the data is hard to explain under H₀ — so you reject it in favor of the alternative hypothesis (H₁).

**4. It enforces intellectual honesty**
Starting from "no effect" forces researchers to meet a burden of proof rather than just finding patterns that confirm what they hoped to find.

---

## The Logic in Brief

```
Assume H₀ is true
        ↓
Collect data
        ↓
Is the data very unlikely under H₀?
   Yes → Reject H₀ (evidence supports H₁)
   No  → Fail to reject H₀ (insufficient evidence)
```

> **Key nuance:** Failing to reject H₀ does *not* prove it's true — it just means you don't have enough evidence against it. This distinction matters a lot in interpreting results.





 What does the significance level (α) represent in hypothesis testing?


 ## Significance Level (α)

The **significance level (α)** is the probability threshold you set *before* running a test that defines how much risk of a **false positive (Type I error)** you're willing to accept — that is, the risk of rejecting H₀ when it's actually true.

In plain terms: **α is the bar the evidence must clear to call a result "statistically significant."**

---

## Common Values

| α Value | Used When |
|--------|-----------|
| **0.05** | Most fields — standard default |
| **0.01** | Medical/clinical research — stricter |
| **0.001** | Physics, genomics — very strict |

---

## How It Works

You choose α *before* collecting data, then compare your p-value to it after:

```
p ≤ α  →  Reject H₀  (result is statistically significant)
p > α  →  Fail to reject H₀  (insufficient evidence)
```

If α = 0.05 and you get p = 0.03, you reject H₀ because 0.03 < 0.05.

---

## What α = 0.05 Actually Means

Setting α = 0.05 means you're accepting a **5% chance** that you'll incorrectly reject a true H₀. In other words, if you ran 100 experiments where H₀ was actually true, you'd expect to get a "significant" result in about 5 of them purely by chance.

---

## The Key Trade-off

Lowering α reduces false positives but increases false negatives (missing real effects):

```
Lower α (e.g., 0.01)          Higher α (e.g., 0.10)
→ Harder to reject H₀         → Easier to reject H₀
→ Fewer false positives        → More false positives
→ More false negatives         → Fewer false negatives
→ More conservative            → More liberal
```

---

## Important Nuances

- **α is not the probability H₀ is true.** It's the long-run error rate of your *decision rule*, not a statement about any single test.
- **α must be set before the test.** Changing it after seeing your p-value ("p-hacking") invalidates the logic entirely.
- **Statistical significance ≠ practical significance.** A result can clear the α threshold yet still be too small an effect to matter in the real world.

## Type I vs Type II Errors

Both errors arise because hypothesis testing is a decision made under uncertainty — you never know with absolute certainty whether H₀ is true or false.

---

## The 2×2 Reality

|  | **H₀ is Actually True** | **H₀ is Actually False** |
|---|---|---|
| **Reject H₀** | ❌ Type I Error (False Positive) | ✅ Correct (True Positive) |
| **Fail to Reject H₀** | ✅ Correct (True Negative) | ❌ Type II Error (False Negative) |

---

## Type I Error (False Positive)

**What it is:** Rejecting H₀ when it is actually true — claiming an effect exists when it doesn't.

**Controlled by:** α (significance level). If α = 0.05, you have a 5% chance of making a Type I error.

**Real-world example:** A drug trial concludes a new drug *does* lower blood pressure, but in reality it has no effect — the result was just random chance.

**Analogy:** Convicting an innocent person.

---

## Type II Error (False Negative)

**What it is:** Failing to reject H₀ when it is actually false — missing a real effect that exists.

**Controlled by:** β (beta). The probability of making a Type II error is β, and **Power = 1 − β** (the probability of correctly detecting a real effect).

**Real-world example:** A drug trial concludes there is *no evidence* the drug works, but the drug actually does lower blood pressure — the sample was just too small to detect it.

**Analogy:** Acquitting a guilty person.

---

## Side-by-Side Comparison

| Feature | Type I Error | Type II Error |
|---|---|---|
| Also called | False Positive | False Negative |
| What happened | Rejected a true H₀ | Failed to reject a false H₀ |
| Probability | α (e.g., 0.05) | β (e.g., 0.20) |
| Reduced by | Lowering α | Increasing sample size, raising α |
| Real-world cost | Acting on a non-existent effect | Missing a real effect |

---

## The Fundamental Trade-off

The two errors pull in **opposite directions** — you can't minimize both simultaneously without increasing sample size:

```
Lower α → Harder to reject H₀
        → Fewer Type I errors ✅
        → More Type II errors ❌

Raise α → Easier to reject H₀
        → More Type I errors ❌
        → Fewer Type II errors ✅
```

The only way to reduce *both* simultaneously is to **collect more data**, which increases the precision of your test.

---

## Which Error is Worse?

It depends entirely on context and consequences:

- **Type I is worse** when false alarms are costly — e.g., approving an ineffective drug, launching an unnecessary policy, or sending innocent people to prison.
- **Type II is worse** when missing a real effect is dangerous — e.g., failing to detect a disease, missing a security threat, or not recalling a faulty product.

This is why medical diagnostics, security screening, and drug safety testing deliberately use very low α — the cost of a false positive is high enough to justify risking more false negatives.

4: Explain the difference between a one-tailed and two-tailed test. Give an example of each.


## One-Tailed vs Two-Tailed Tests

The "tail" refers to the **region of rejection** in the probability distribution — the area where, if your test statistic lands, you reject H₀. The choice depends entirely on the **direction of your hypothesis**.

---

## Visual Intuition

```
Two-Tailed (α split both sides)        One-Tailed (α on one side only)

  Reject    Accept     Reject             Accept        Reject
  ←2.5% |           | 2.5%→             ←          | 5%→
  ───────────────────────────            ──────────────────────
       -1.96       +1.96                           +1.645
```

---

## Two-Tailed Test

**When to use it:** You want to detect a difference in **either direction** — the effect could be larger or smaller than the baseline.

**Hypotheses:**
- H₀: There is no difference (μ = μ₀)
- H₁: There is a difference in either direction (μ ≠ μ₀)

**How α is applied:** Split equally across both tails. At α = 0.05, each tail gets 2.5%.

**Example:**
> A school wants to know if a new teaching method changes exam scores — it could improve *or* worsen them.
> - H₀: Mean score = 70
> - H₁: Mean score ≠ 70
>
> A result far above **or** far below 70 would lead to rejecting H₀.

---

## One-Tailed Test

**When to use it:** You have a specific directional prediction — the effect can only be meaningfully interpreted in **one direction.**

**Two variants:**

**Right-tailed (upper):**
- H₁: μ > μ₀ — you expect an *increase*
- Rejection region is on the **right**

**Left-tailed (lower):**
- H₁: μ < μ₀ — you expect a *decrease*
- Rejection region is on the **left**

**Example (Right-tailed):**
> A company claims its new battery lasts *longer* than the current 10-hour average.
> - H₀: Mean battery life ≤ 10 hours
> - H₁: Mean battery life > 10 hours
>
> Only an unusually **high** result would reject H₀ — a lower result is irrelevant to the claim.

**Example (Left-tailed):**
> A manufacturer checks if a new process *reduces* defect rates below the current 8%.
> - H₀: Defect rate ≥ 8%
> - H₁: Defect rate < 8%
>
> Only an unusually **low** defect rate would reject H₀.

---

## Side-by-Side Comparison

| Feature | Two-Tailed | One-Tailed |
|---|---|---|
| Direction of H₁ | ≠ (either direction) | > or < (one direction) |
| Rejection regions | Both tails | One tail only |
| α distribution | Split: α/2 each side | All α on one side |
| Critical value (α=0.05) | ±1.96 | 1.645 (right) or −1.645 (left) |
| More conservative? | ✅ Yes | ❌ No |
| When to use | No prior directional theory | Strong prior directional prediction |

---

## The Critical Trade-off

Because a one-tailed test puts **all** of α on one side, it is more sensitive to detecting an effect in that direction — but **completely blind** to effects in the opposite direction.

```
Same data, α = 0.05, p = 0.04 (one-tailed):
  → One-tailed test: Reject H₀ ✅
  → Two-tailed test: Fail to reject H₀ ❌ (would need p < 0.025)
```

---

## A Word of Caution

One-tailed tests must be chosen **before** seeing the data, based on genuine theoretical reasoning. Switching to a one-tailed test *after* seeing results lean in one direction is a form of p-hacking — it artificially inflates your chance of finding significance and undermines the integrity of the test. When in doubt, **default to two-tailed** — it's the more conservative and defensible choice.

  ## Z-test vs T-test: When to Use Which

The core question is: **how much do you know about the population, and how large is your sample?**

---

## The Key Decision Criteria

| Condition | Z-test | T-test |
|---|---|---|
| Population standard deviation (σ) | **Known** | **Unknown** (use sample SD) |
| Sample size | **Large (n ≥ 30)** | **Small (n < 30)** |
| Population distribution | Normal (or CLT applies) | Normal (or approximately normal) |
| Proportions testing | ✅ Yes | ❌ Rarely |

---

## Why the Distinction Exists

When σ is **unknown**, you estimate it using the sample standard deviation (s). This estimation introduces **extra uncertainty** — especially problematic with small samples. The t-distribution accounts for this by having **heavier tails** than the normal distribution, requiring stronger evidence to reject H₀.

```
Z-distribution:                 T-distribution (small n):
    ___                               ___
   /   \                             /   \
  /     \                           /     \
_/       \_____              ______/       \______
  Thin tails                    Heavier tails
  (less uncertainty)            (more uncertainty)
```

As sample size grows, the t-distribution gradually converges toward the Z-distribution. At n ≥ 30, they are practically indistinguishable.

---

## Use a Z-test When

**1. Population standard deviation (σ) is known**
> Rare in practice, but occurs when historical data or regulatory standards define σ precisely.
> Example: A factory knows from years of records that fill volume has σ = 2ml. Testing if a new machine changes the mean → Z-test.

**2. Large sample size (n ≥ 30), σ unknown**
> By the Central Limit Theorem, the sampling distribution of x̄ is approximately normal for large n. The sample SD (s) becomes a reliable estimate of σ.
> Example: Testing whether 50 students' average score differs from a national benchmark → Z-test acceptable.

**3. Testing proportions**
> For testing population proportions (p), the Z-test is standard — provided np ≥ 5 and n(1−p) ≥ 5.
> Example: Testing if more than 40% of customers prefer a new product → Z-test.

---

## Use a T-test When

**1. Population standard deviation (σ) is unknown AND n < 30**
> The most common real-world scenario. You only have sample data and a small group.
> Example: Testing a new drug on 15 patients → T-test.

**2. Small sample from a normal population**
> Even if you could argue CLT applies loosely, t-test is the safer, more honest choice.
> Example: Comparing average battery life across 12 devices → T-test.

**3. Comparing two small sample means**
> Two-sample t-test handles comparisons between two groups with unknown σ.
> Example: Comparing exam scores of two classes of 20 students each → T-test.

---

## Decision Flowchart

```
Start: What are you testing?
            │
            ▼
     Testing a proportion?
       Yes → Z-test
       No  ↓
            ▼
    Is σ (population SD) known?
       Yes → Z-test
       No  ↓
            ▼
      Is n ≥ 30?
       Yes → Z-test (CLT justifies it)
       No  ↓
            ▼
      Is population roughly normal?
       Yes → T-test
       No  → Consider non-parametric tests
```

---

## Side-by-Side Formula Comparison

**Z-test statistic:**
$$Z = \frac{\bar{x} - \mu_0}{\sigma / \sqrt{n}}$$
Uses **population** standard deviation σ

**T-test statistic:**
$$t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}$$
Uses **sample** standard deviation s — referenced against t-distribution with df = n − 1

The formulas look identical except for σ vs s — but the **reference distribution** used to find the critical value is fundamentally different.

---

## Practical Reality

In most real-world research, **σ is almost never known** — which means the t-test is the default for means testing. The Z-test for means is largely a textbook tool that illustrates concepts, while the t-test is what practitioners actually use. The Z-test remains genuinely useful primarily for **proportion testing** and **very large samples**.

## Chi-Square Test of Independence: Gender vs Product Preference

---

## Step 1: State the Hypotheses

- **H₀:** Product preference is **independent** of gender
- **H₁:** Product preference is **not independent** of gender (there is an association)

---

## Step 2: Observed Frequencies (O)

| Gender | Product A | Product B | Total |
|---|---|---|---|
| **Male** | 30 | 20 | 50 |
| **Female** | 10 | 40 | 50 |
| **Total** | 40 | 60 | 100 |

---

## Step 3: Calculate Expected Frequencies (E)

Formula for each cell:

$$E = \frac{(\text{Row Total}) \times (\text{Column Total})}{\text{Grand Total}}$$

| Cell | Calculation | Expected (E) |
|---|---|---|
| Male, Product A | (50 × 40) / 100 | **20** |
| Male, Product B | (50 × 60) / 100 | **30** |
| Female, Product A | (50 × 40) / 100 | **20** |
| Female, Product B | (50 × 60) / 100 | **30** |

**Expected Frequencies Table:**

| Gender | Product A | Product B | Total |
|---|---|---|---|
| **Male** | 20 | 30 | 50 |
| **Female** | 20 | 30 | 50 |
| **Total** | 40 | 60 | 100 |

> ✅ All expected frequencies > 5 — Chi-square test is valid to proceed.

---

## Step 4: Calculate the Chi-Square Statistic

$$\chi^2 = \sum \frac{(O - E)^2}{E}$$

| Cell | O | E | O − E | (O − E)² | (O − E)²/E |
|---|---|---|---|---|---|
| Male, Product A | 30 | 20 | +10 | 100 | **5.00** |
| Male, Product B | 20 | 30 | −10 | 100 | **3.33** |
| Female, Product A | 10 | 20 | −10 | 100 | **5.00** |
| Female, Product B | 40 | 30 | +10 | 100 | **3.33** |
| | | | | **χ²** | **16.66** |

$$\chi^2 = 5.00 + 3.33 + 5.00 + 3.33 = 16.66$$

---

## Step 5: Find the Critical Value

- Degrees of freedom: df = (rows − 1)(columns − 1) = (2−1)(2−1) = **1**
- Significance level: α = 0.05
- From Chi-square table: **χ²_critical = 3.841**

```
      Fail to Reject H₀          Reject H₀
←—————————————————————|————————————→
            0        3.841
                            χ² = 16.66 ↑
                          (falls in rejection region)
```

---

## Step 6: Decision

| | Value |
|---|---|
| Calculated χ² | 16.66 |
| Critical χ² | 3.841 |
| χ² > χ²_critical? | 16.66 > 3.841? **Yes** ✅ |

Since **16.66 > 3.841**, the test statistic falls **in the rejection region**.

→ **Reject H₀**

---

## Step 7: Conclusion

> At α = 0.05, there is **sufficient statistical evidence** to conclude that product preference is **not independent of gender** — a significant association exists between gender and product preference.

Looking at the data more closely:
